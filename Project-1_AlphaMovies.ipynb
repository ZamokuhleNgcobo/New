```python
#Import packages

import pandas as pd
import numpy as np
import random
```

```python

#Data Importation
movie_data = pd.read_csv("/Users/dam154/Downloads/imdb_top_1000.csv")

movie_data.head()
```

```python


#Column Deletion
#el movie_data['Poster_Link']       #Delete the certificate column
del movie_data['Certificate']       #Delete the certificate column
del movie_data['No_of_Votes']       #Delete the Number of votes
del movie_data['Gross']             #Delete the Gross annual return column
del movie_data['Meta_score']        #Delete the Meta Score column

#display data
movie_data.head(5)
```

Creating a variable representing the number of copies available in the Store

```python
#Create a 1000 ranndom numbers for available copies
rand_list = []

for i in range(0,1000):
    n = random.randint(1,50)
    rand_list.append(n)
print(rand_list)



```

```python
#Convert the list to a Dataframe

copies_available = pd.DataFrame(rand_list)
copies_available.head(20)

#Getting the length of the dataframe
len(copies_available)

copies_available.head(5)
```

```python
#Create a column of numbers on both datasets

movie_data = movie_data.reset_index()
movie_data.head(5)
```

```python

#Create a column of index numbers on the new dataset

copies_available = copies_available.reset_index()
copies_available.head(5)
```

```python
#Merge the two the dataframes
movies = pd.merge(movie_data,copies_available)

#Rename necessary columns
movies.rename(columns= {0:'Copies_available', 'IMDB_Rating':'Rating','Series_Title' : 'Movies'}, inplace = True)
movies.head(10)
```

Create a variable(column) representing price of rental of the movie

```python
#Create a list categories based on ratings

movies['Prices'] = 20 * movies['Rating']

movies.head(20)
```
